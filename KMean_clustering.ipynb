```python
import pandas as pd
data = pd.read_csv("/content/gdrive/MyDrive/Kaggle_dataset/heart.csv")
data
```

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
data.head()
```

```python
plt.scatter(data.Age,data.MaxHR)
plt.xlabel("Age")
plt.ylabel("MAXIMUM HEART RATE")
```

```python
km = KMeans(n_clusters = 10)
predicted = km.fit_predict(data[["Age" , "MaxHR"]])
predicted
```

```python
data["cluster"] = predicted
data.head()
data1 = data[data.cluster == 0]
data2 = data[data.cluster == 1]
data3 = data[data.cluster == 2]
data4 = data[data.cluster == 3]
data5 = data[data.cluster == 4]
data6 = data[data.cluster == 5]
data7 = data[data.cluster == 6]
data8 = data[data.cluster == 7]
data9 = data[data.cluster == 8]
data10 = data[data.cluster == 9]
plt.scatter(data1.Age , data1["MaxHR"],color = "red")
plt.scatter(data2.Age , data2["MaxHR"],color = "blue")
plt.scatter(data3.Age , data3["MaxHR"],color = "green")
plt.scatter(data4.Age , data4["MaxHR"],color = "orange")
plt.scatter(data5.Age , data5["MaxHR"],color = "pink")
plt.scatter(data6.Age , data6["MaxHR"],color = "gray")
plt.scatter(data7.Age , data7["MaxHR"],color = "cyan")
plt.scatter(data8.Age , data8["MaxHR"],color = "purple")
plt.scatter(data9.Age , data9["MaxHR"],color = "lime")
plt.scatter(data10.Age , data10["MaxHR"],color = "brown")
plt.xlabel("Age")
plt.ylabel("MAXHR")
```
